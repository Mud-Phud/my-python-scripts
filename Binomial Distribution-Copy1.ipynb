In this page we look at the binomialdistribution. If we assume equal probabilities of flipping a coin, and we take $n$ flips, then the probability of ${n \choose j}\frac{1}{2^n}$ for $j=1,\ldots, n+1$. The distribution is can plotted as follows:

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

n =10
p = 0.5


x_pos = [i for i in range(n+1)]

# we need the choose function...

def choose(n, k):
    """
    A fast way to calculate binomial coefficients by Andrew Dalke (contrib).
    """
    if 0 <= k <= n:
        ntok = 1
        ktok = 1
        for t in range(1, min(k, n - k) + 1):
            ntok *= n
            ktok *= t
            n -= 1
        return ntok // ktok
    else:
        return 0
    
pdf = [ choose(n, i)*p**i*(1-p)**(n-i) for i in range(n+1)]

plt.bar(x_pos, pdf, color='green')
plt.xlabel("Number of heads")
plt.ylabel("Probability")
plt.title("Binomial distribution for n = "+str(n)+", p = "+str(p))

plt.xticks(x_pos)

plt.show()
```
