$$
\tau(s)= \int\limits _{0}^{\infty}\frac{x^{s-1}}{\left(e^{x}-1\right)}dx
$$

Let us look at the integrand for various values of $s$.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))

fig, ax = plt.subplots(3, 
           sharex='col', sharey='row')
x = np.linspace(0.2,5, 100)
s1 = 1.2
y1 = x**(s1-1)/(np.exp(x)-1)
ax[0].plot(x, y1)
ax[0].set(title='$s=1.2$')

s2 = 1.5
y2 = x**(s2-1)/(np.exp(x)-1)
ax[1].plot(x, y2)
ax[1].set(title="$s=1.5$")

s3 = 2.5
y3 = x**(s2-1)/(np.exp(x)-1)
ax[2].plot(x, y3)
ax[2].set(title="$s=2.5$")

plt.show()
```

Now let's look the original function 
$$
\tau(s)= \int\limits _{0}^{\infty}\frac{x^{s-1}}{\left(e^{x}-1\right)}dx
$$
We will do this by using Simpson's integration

```python

def Simp(f,a,b,n):
    if n % 2 == 0: n += 1  # n needs to be odd.
    h = (b-a)/(n-1)

    x = np.linspace(a,b,n) 
    y = f(x)
    return (h/3)*(2*sum(y[::2])-y[0]-y[-1] + 4*sum(y[1::2]))

s_range = np.linspace(1.02,4,50)
y = []

for s in s_range:
    def f(x):
        return x**(s-1)/(np.exp(x)-1)
    y.append(Simp(f,0.01,10,100))


plt.rcParams['figure.figsize'] = [10, 5]
plt.plot(s_range, y, color='r')
plt.title("Integral defined function")

plt.show()
```

This blows up at $s=0$. Let's add an $s-1$ factor to the integrand:

```python

s_range = np.linspace(1.02,2,50)
y = []

for s in s_range:
    def f(x):
        return (s-1)*x**(s-1)/(np.exp(x)-1)
    y.append(Simp(f,0.001,10,1000))


plt.rcParams['figure.figsize'] = [10, 5]
plt.plot(s_range, y, color='r')
plt.title("Integral defined function")

plt.show()
```
